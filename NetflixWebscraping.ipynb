```python
#using DataFrame
Netflix_data=pd.DataFrame(columns=['Date','Open','High','Low','Close','Volume'])

for row in soup.find('tbody').find_all('tr'):
        col=row.find_all('td')
        Date=col[0].text
        Open=col[1].text
        High=col[2].text
        Low=col[3].text
        Close=col[4].text
        Volume=col[5].text
        Netflix_data=Netflix_data.append({"Date":Date,"Open":Open,'High':High,'Low':Low, 'Close':Close,'Volume':Volume}, ignore_index=True)

Netflix_data
```

```python
#Neflix Revenue

url="https://www.macrotrends.net/stocks/charts/NFLX/netflix/revenue"

ndata=requests.get(url).text
```

```python
bsoup=BeautifulSoup(ndata,'html5lib')
```

```python
df=pd.DataFrame(columns=['Date','Amount in USD'])

for i in bsoup.find_all('table'):
    if("Netflix Quarterly Revenue" in i.find('th').text):

                row=i.find_all("tr")
                
                for r in row:
                    col=r.find_all("td")
                    if col!=[]:
                        date=col[0].text
                        amount=col[1].text
                        df=df.append({'Date':date,'Amount in USD':amount},ignore_index=True)
                
df
       
```
