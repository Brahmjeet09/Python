```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url="https://en.wikipedia.org/wiki/Sikh_gurus"

data=requests.get(url).text
soup=BeautifulSoup(data,'html5lib')
```

```python
Gurus_df=pd.DataFrame(columns=['Name', 'Birth Date', 'Birth Place','Date of Death'])
heading=soup.find('tbody')

for h in heading.find_all('tr'):
        col=h.find_all('td')
        if col!=[]:
            name=col[1].text
            bdate=col[4].text
            bplace=col[5].text
            dod=col[9].text
            
            Gurus_df=Gurus_df.append({'Name':name,'Birth Date':bdate,'Birth Place':bplace,'Date of Death':dod},ignore_index=True)


Gurus_df["Name"] = Gurus_df['Name'].str.replace('\n',"")
Gurus_df["Birth Date"] = Gurus_df['Birth Date'].str.replace('\n|[note 1]',"")
Gurus_df["Birth Place"] = Gurus_df['Birth Place'].str.replace('\n',"")
Gurus_df["Date of Death"] = Gurus_df['Date of Death'].str.replace('\n',"")
Gurus_df
```
